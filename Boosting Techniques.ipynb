# 25.Train an AdaBoost Classifier with different numbers of estimators and compare accuracy

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
```

```python
# load the datasets
cancer = load_breast_cancer()
x = cancer.data
y = cancer.target
```

```python
# split and train and test the data
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
#   different numbers of estimators
estimators = [10, 50, 100, 200]
accuracy = []
```

```python
#   different numbers of estimators
estimators = [10, 50, 100, 200]
# Re-initialize accuracy as an empty list
accuracy = []

# train a model using for loop
for n in estimators:
    model = AdaBoostClassifier(n_estimators=n, random_state=42)
    model.fit(x_train, y_train)
    y_pred = model.predict(x_test)
    acc = accuracy_score(y_test, y_pred)
    accuracy.append(acc)
    print(f"n_estimators = {n}, Accuracy = {acc:.4f}")
```

```python
# plotting the map
plt.plot(estimators, accuracy)
plt.xlabel('n_estimators')
plt.ylabel('Accuracy')
plt.title('Accuracy vs n_estimators')
plt.show()
```

# 26.Train a Gradient Boosting Classifier and visualize the ROC curve.

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt
```

```python
# load the datasets
cancer = load_breast_cancer()
x = cancer.data
y = cancer.target
```

```python
# split train and test data
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
# train with the model
model = GradientBoostingClassifier(n_estimators=100, random_state=42)
model.fit(x_train, y_train)
```

```python
# predict the model
y_pred = model.predict_proba(x_test)[:, 1]
```

```python
# check rou
fpr, tpr, thresholds = roc_curve(y_test, y_pred)
roc_auc = auc(fpr, tpr)
```

```python
# print the values of it
print("False Positive Rate:", fpr)
print("True Positive Rate:", tpr)
print("Thresholds:", thresholds)
print("ROC AUC:", roc_auc)
```

```python
# plotting the map
plt.plot(fpr, tpr, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], 'k--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.legend(loc="lower right")
plt.show()
```

# 27.Train an XGBoost Regressor and tune the learning rate using GridSearchCV.

```python
from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split, GridSearchCV
from xgboost import XGBRegressor
```

```python
# load the data set
diabetes = load_diabetes()
x = diabetes.data
y = diabetes.target
```

```python
# split train and test
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
# define the model
model = XGBRegressor(n_estimators=100, random_state=42)
```

```python
# learning model
param_grid = {'learning_rate': [0.01, 0.1, 0.5, 1.0]}
grid_search = GridSearchCV(model, param_grid, cv=5)
```

```python
# train the model
grid_search.fit(x_train, y_train)
```

```python
# predict the model
y_pred = grid_search.predict(x_test)
```

```python
# Best parameters
print("Best learning rate:", grid_search.best_params_['learning_rate'])
```

```python
# Evaluate on test data
import numpy as np
from sklearn.metrics import mean_squared_error
best_model = grid_search.best_estimator_
y_pred = best_model.predict(x_test)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
print(f"Test RMSE: {rmse:.4f}")
```

# 28.Train a CatBoost Classifier on an imbalanced dataset and compare performance with class weighting.

```python
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from catboost import CatBoostClassifier
```

```python
# load the imbalanced data
x, y = make_classification(n_samples=1000, n_features=20, weights=[0.9, 0.1], random_state=42)
```

```python
# split the data
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
# train the model
model = CatBoostClassifier(iterations=100, learning_rate=0.1, depth=6, loss_function='Logloss')
model.fit(x_train, y_train)
```

```python
# predict the model
y_pred = model.predict(x_test)
```

```python
# Model WITHOUT class weights
model_no_weights = CatBoostClassifier(verbose=0, random_state=42)
model_no_weights.fit(x_train, y_train)
preds_no_weights = model_no_weights.predict(x_test)
```

```python
# campare for the model
accuracy_no_weights = accuracy_score(y_test, preds_no_weights)
accuracy_with_weights = accuracy_score(y_test, y_pred)
```

```python
# print the values of both the model
print("Accuracy without class weights:", accuracy_no_weights)
print("Accuracy with class weights:", accuracy_with_weights)
```

# 29.Train an AdaBoost Classifier and analyze the effect of different learning rates

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
# load the datasets
iris = load_iris()
x = iris.data
y = iris.target
```

```python
# split the train and test
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
# Step 4: Try different learning rates
learning_rates = [0.01, 0.1, 0.5, 1.0, 1.5, 2.0]
accuracy_scores = []
```

```python
# using for loop for the data
for learning_rate in learning_rates:
    model = AdaBoostClassifier(n_estimators=100, random_state=0, learning_rate=learning_rate)
    model.fit(x_train, y_train)
    y_pred = model.predict(x_test)
    accuracy = accuracy_score(y_test, y_pred)
    accuracy_scores.append(accuracy)
    print("Accuracy for learning rate {}: {}".format(learning_rate, accuracy))
```

```python
# load the datasets
iris = load_iris()
x = iris.data
y = iris.target

# split the train and test
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

# Step 4: Try different learning rates
learning_rates = [0.01, 0.1, 0.5, 1.0, 1.5, 2.0]
accuracy_scores = [] # Initialize the list before the loop

# using for loop for the data
for learning_rate in learning_rates:
    model = AdaBoostClassifier(n_estimators=100, random_state=0, learning_rate=learning_rate)
    model.fit(x_train, y_train)
    y_pred = model.predict(x_test)
    accuracy = accuracy_score(y_test, y_pred)
    accuracy_scores.append(accuracy) # Append only once per learning rate
    print("Accuracy for learning rate {}: {}".format(learning_rate, accuracy))

# insert the plot
import matplotlib.pyplot as plt
plt.plot(learning_rates, accuracy_scores, marker='o')
plt.xlabel('Learning Rate')
plt.ylabel('Accuracy')
plt.title('Effect of Learning Rate on Accuracy')
plt.grid(True)
plt.show()
```
